```python
def scrape_ieee_articles(query, num_pages):
    chrome_service = Service(executable_path="C:\\Users\\jissr\\chromedriver.exe")  # Update with your path
    options = Options()
    options.add_argument('--headless')
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')

    driver = webdriver.Chrome(service=chrome_service, options=options)

    articles = []
    page = 1

    while page <= num_pages:
        url = f"https://ieeexplore.ieee.org/search/searchresult.jsp?queryText={query}&highlight=true&returnType=SEARCH&rowsPerPage=25&pageNumber={page}"
        driver.get(url)

        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, 'List-results-items'))
        )

        soup = BeautifulSoup(driver.page_source, "html.parser")

        results = soup.find_all("div", class_="List-results-items")

        if not results:
            print(f"No results found on page {page}.")
            break

        for result in results:
            # Extract title
            title_tag = result.find("a", class_="fw-bold")
            title = title_tag.text.strip() if title_tag else "No title available"

            # Extract document URL
            doc_url = title_tag['href'] if title_tag and 'href' in title_tag.attrs else "No URL available"

            # Extract authors
            authors_tag = result.find("p", class_="author")
            authors = authors_tag.text.strip() if authors_tag else "N/A"

            # Extract publication year and conference/journal
            pub_info_tag = result.find("div", class_="description")
            pub_info = pub_info_tag.text.strip() if pub_info_tag else "No Publication Info"
            year = None
            if "Year:" in pub_info:
                year = pub_info.split("Year: ")[1].split()[0]
            conference_or_journal = pub_info.split(",")[0] if pub_info else "N/A"

            # Extract citation count
            citation_count = "N/A"
            citation_tags = result.find_all("div", class_="cited-by")  # Update with correct class
            for citation_tag in citation_tags:
                citation_text = citation_tag.text.strip()
                if "Cited by:" in citation_text:
                    # Extract the numerical part
                    citation_count = citation_text.split("Cited by: ")[1].split()[0]
                    citation_count = ''.join(filter(str.isdigit, citation_count))  # Keep only digits

            # Append extracted information to the articles list
            articles.append({
                "Title": title,
                "Authors": authors,
                "Year": year,
                "Conference/Journal": conference_or_journal,
                "Document URL": f"https://ieeexplore.ieee.org{doc_url}",
                "Citation Count": citation_count if citation_count else "N/A",
            })

        page += 1
        time.sleep(3)  # Add delay to avoid being blocked

    driver.quit()
    return articles

def save_to_excel(articles, filename):
    df = pd.DataFrame(articles)
    df.to_excel(filename, index=False)
    df.head()

# Example usage
query = "artificial intelligence"
num_pages = 1  # Adjust as needed
articles = scrape_ieee_articles(query, num_pages)
save_to_excel(articles, "ieee_articles.xlsx")
print(f"Saved {len(articles)} articles to ieee_articles.xlsx")
```
